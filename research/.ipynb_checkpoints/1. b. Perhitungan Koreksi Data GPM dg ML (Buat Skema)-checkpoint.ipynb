## Menentukan features & target

```python
variables = {"Elevasi": True,
            "Jarak": True,  
            "latitude": True, 
            "longitude": True,
            "T": True,
            "pu_gpm": True}

features = []
for key,val in variables.items():
    if val == True:
        features.append(key)
        
print(f"features : {features}")

target = "pu_station"

# Separate features (X) and target variable (y) for test and train datasets
test_datasets = [test_1, test_2, test_3, test_4, test_5]
train_datasets = [train_1, train_2, train_3, train_4, train_5]
```

```python
#test_datasets[4]
```

```python
#from xgboost import XGBRegressor
```

```python
all_X_train_data = []
all_y_train_data = []
all_X_test_data = []
all_y_test_data = []

# Initialize lists to store predictions
all_test_predictions = []

for idx, (X_train, y_train, X_test, y_test) in enumerate(zip(train_datasets, [train_1, train_2, train_3, train_4, train_5], test_datasets, [test_1, test_2, test_3, test_4, test_5]), start=1):
    # Separate features and target variable for train dataset
    X_train_data = X_train[features]
    y_train_data = X_train[target]

    # Separate features and target variable for test dataset
    X_test_data = X_test[features]
    y_test_data = X_test[target]
    
    # Append to the list
    all_X_train_data.append(X_train_data)
    all_y_train_data.append(y_train_data)
    all_X_test_data.append(X_test_data)
    all_y_test_data.append(y_test_data)

    # Initialize the regression model
    #model = LinearRegression()
    #model = XGBRegressor()
    model = RandomForestRegressor(n_estimators=67, random_state=95)

    # Train the model on the current train dataset
    model.fit(X_train_data, y_train_data)

    # Make predictions on the current test dataset
    y_pred = model.predict(X_test_data)

    # Save predictions to the list
    test_predictions = pd.DataFrame({'pu_true': y_test_data, 'pu_pred': y_pred})
    all_test_predictions.append(test_predictions)

    # Print or use the predictions and actual values as needed
    print(f"Predictions for Test_{idx}:")
    print(test_predictions)

#Concenate all test and train to dataframe
df_all_X_train_data = pd.concat(all_X_train_data, ignore_index=False)
df_all_y_train_data = pd.concat(all_y_train_data, ignore_index=False)
df_all_X_test_data = pd.concat(all_X_test_data, ignore_index=False)
df_all_y_test_data = pd.concat(all_y_test_data, ignore_index=False)
    
# Concatenate all predictions into one DataFrame
df_all_predictions = pd.concat(all_test_predictions, keys=[f'Test_{idx}' for idx in range(1, 6)])

# Display the concatenated predictions
##print("All Test Predictions:")
##print(df_all_predictions)
```

## Evaluasi Model

```python
class Evaluation:
    def __init__(self, model):
        self.model = model

    def calculate_mse(self, y_pred, y_true):
        mse = mean_squared_error(y_pred=y_pred, y_true= y_true)
        return mse

    def calculate_rmse(self, y_pred, y_true):
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        return rmse
    
    def calculate_nmse_mean(self, y_pred, y_true):
        rmse = self.calculate_rmse(y_pred=y_pred , y_true=y_true)
        nmse_mean = rmse/np.mean(y_true)
        return nmse_mean
    
    def calculate_nmse_std(self, y_pred, y_true):
        rmse = self.calculate_rmse(y_pred=y_pred , y_true=y_true)
        nmse_std = rmse/np.std(y_true)
        return nmse_std
    
    def calculate_nmse_min_max(self, y_pred, y_true):
        rmse = self.calculate_rmse(y_pred=y_pred, y_true=y_true)
    
        # Menangani kasus ketika nilai maksimum dan minimum sama
        if np.max(y_true) == np.min(y_true):
        # Handle secara khusus untuk menghindari pembagian oleh nol
            nmse_min_max = 0.0
        else:
            nmse_min_max = rmse / (np.max(y_true) - np.min(y_true))
    
        return nmse_min_max
        
evaluation = Evaluation(model)
```

```python
nmse_all = []
for i in range(2,20):
    Xi = df_all_X_test_data[df_all_X_test_data['T'] == i]
    nmse = evaluation.calculate_nmse_min_max(Xi['pu_pred'].values, Xi['pu_true'].values)
    nmse_all.append(nmse)
    print(f"NMSE periode ulang {i} adalah {nmse}")
```

## NMSE Mean (Setelah Koreksi)

```python
unique_periods = df_all_X_test_data['T'].unique()
```

```python
unique_periods
```

```python
# Inisialisasi list untuk menyimpan NMSE untuk setiap baris
nmse_all_after = []

# Loop untuk setiap periode ulang
for period in unique_periods:
    # Buat subset data untuk periode ulang tertentu
    Xi = df_all_X_test_data[df_all_X_test_data['T'] == period]
    
    # Loop untuk setiap baris dalam periode ulang
    for index, row in Xi.iterrows():
        # Buat DataFrame kecil untuk setiap baris
        single_row_df = pd.DataFrame([row])
        
        # Hitung NMSE untuk setiap baris
        nmse_after = evaluation.calculate_nmse_mean(single_row_df['pu_pred'].values, single_row_df['pu_true'].values)
        
        # Tambahkan hasil NMSE ke list sebagai tuple
        nmse_all_after.append((period, index, nmse_after))
```

```python
# Cetak hasil NMSE
for result in nmse_all_after:
    print(f"NMSE untuk periode ulang {result[0]}, baris {result[1]} adalah {result[2]}")
```

```python
# Pilih periode ulang yang ingin ditampilkan NMSE-nya
target_period = 3  # Ganti dengan periode ulang yang diinginkan

# Cetak hasil NMSE hanya untuk periode ulang tertentu
for result in nmse_all_after:
    if result[0] == target_period:
        print(f"NMSE untuk periode ulang {result[0]}, baris {result[1]} adalah {result[2]}")
```

## NMSE Mean (Sebelum Koreksi)

```python
# Inisialisasi list untuk menyimpan NMSE untuk setiap baris
nmse_all_before = []

# Loop untuk setiap periode ulang
for period in unique_periods:
    # Buat subset data untuk periode ulang tertentu
    Xi = df_all_X_test_data[df_all_X_test_data['T'] == period]
    
    # Loop untuk setiap baris dalam periode ulang
    for index, row in Xi.iterrows():
        # Buat DataFrame kecil untuk setiap baris
        single_row_df = pd.DataFrame([row])
        
        # Hitung NMSE untuk setiap baris
        nmse_before = evaluation.calculate_nmse_mean(single_row_df['pu_gpm'].values, single_row_df['pu_true'].values)
        
        # Tambahkan hasil NMSE ke list sebagai tuple
        nmse_all_before.append((period, index, nmse_before))
```

```python
# Cetak hasil NMSE
for result in nmse_all_before:
    print(f"NMSE untuk periode ulang {result[0]}, baris {result[1]} adalah {result[2]}")
```

```python
# Pilih periode ulang yang ingin ditampilkan NMSE-nya
target_period = 3  # Ganti dengan periode ulang yang diinginkan

# Cetak hasil NMSE hanya untuk periode ulang tertentu
for result in nmse_all_before:
    if result[0] == target_period:
        print(f"NMSE untuk periode ulang {result[0]}, baris {result[1]} adalah {result[2]}")
```

## Selisih NRMSE Sesudah - Sebelum Koreksi

```python
# Pastikan kedua list memiliki jumlah elemen yang sama
if len(nmse_all_before) != len(nmse_all_after):
    raise ValueError("Jumlah elemen dalam kedua list harus sama")

# Inisialisasi list untuk menyimpan hasil pengurangan NMSE
nmse_difference = []

# Loop untuk setiap elemen (baris) dalam list
for i in range(len(nmse_all_before)):
    # Ambil nilai NMSE dari kedua list
    nmse_before = nmse_all_before[i][2]
    nmse_after = nmse_all_after[i][2]
    
    # Hitung selisih NMSE
    nmse_diff = nmse_after - nmse_before
    
    # Tambahkan hasil selisih ke list baru
    nmse_difference.append((nmse_all_before[i][0], nmse_all_before[i][1], nmse_diff))

# Cetak hasil selisih NMSE
for result in nmse_difference:
    print(f"Selisih NMSE untuk periode ulang {result[0]}, baris {result[1]} adalah {result[2]}")
```

```python
# Pilih periode ulang yang ingin ditampilkan NMSE-nya
target_period = 3  # Ganti dengan periode ulang yang diinginkan

# Cetak hasil NMSE hanya untuk periode ulang tertentu
for result in nmse_difference:
    if result[0] == target_period:
        print(f"NMSE untuk periode ulang {result[0]}, baris {result[1]} adalah {result[2]}")
```

## Plot Boxplot untuk Nilai NRMSE Seluruh Stasiun (108 stasiun)

```python
import matplotlib.pyplot as plt

# Ganti nmse_all dengan nmse_all_before atau nmse_difference sesuai kebutuhan
data_to_visualize = nmse_difference  # atau nmse_difference

# Extract data for visualization
periods_to_plot = [2, 5, 10, 25, 50, 100]

# Inisialisasi dictionary untuk menyimpan nilai NMSE untuk setiap periode ulang
nmse_by_period = {period: [] for period in periods_to_plot}

# Populate dictionary dengan nilai NMSE sesuai periode ulang
for result in data_to_visualize:
    period = result[0]
    if period in periods_to_plot:
        nmse_by_period[period].append(result[2])

# Buat boxplot
fig, ax = plt.subplots()
ax.boxplot(nmse_by_period.values(), labels=nmse_by_period.keys())#, showfliers=False)
ax.set_xlabel('Return Period (Yr)')
ax.set_ylabel('NMSE Mean')
ax.set_title('Difference NMSE Mean (Simulasi A)')

# Tampilkan plot
plt.show()
```
